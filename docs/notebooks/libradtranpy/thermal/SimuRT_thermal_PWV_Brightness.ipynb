```python
am=1.
pressure = 0.
pwv=4.0
aer=0
ozone=300.
cloudext=0
```

```python
pwvs=np.arange(0.,11.,1.)
```

```python
NBOBS=len(pwvs)
```

```python
jet = plt.get_cmap('jet')
cNorm = colors.Normalize(vmin=0, vmax=NBOBS)
scalarMap = cmx.ScalarMappable(norm=cNorm, cmap=jet)
all_colors = scalarMap.to_rgba(np.arange( NBOBS), alpha=1)
```

```python
all_temp = []
for index in np.arange(NBOBS):

    path,thefile=libsimulateThermal.ProcessSimulation(am,pwvs[index],ozone,pressure,
                                                                       prof_str='us',proc_str='sa',cloudext=cloudext,thermal_output='brightness')
    data = np.loadtxt(os.path.join(path,thefile))
    wl = data[:,0]
    atm = data[:,1] 
    all_temp.append(atm)
```

```python
wl1=wl
```

```python
for index,pwv in enumerate(pwvs): 
    label = f"pwv ={pwv:.1f} mm"
    if index != 0:
        plt.plot(wl1,all_temp[index],'-',color=all_colors[index],label=label)

plt.xlim(0,20000)
plt.grid()
plt.title("Brightness for pwv from 0.2 mm to 12 mm, no cloud")
plt.xlabel("$\\lambda$ (nm)")
plt.ylabel("Brightness ($K$)")
plt.legend()
figname=string_date+"_multibrightness_pwv_cld0.png"    

plt.savefig(figname)
```

```python
List_Of_All_sites = list(Dict_Of_sitesPressures.keys()) 
List_Of_All_sites 
```

```python
NSites = len(List_Of_All_sites)
```

```python
hls_colors = sns.color_palette("hls", NSites)
sns.palplot(hls_colors)
```

```python
import sys
sys.path
```

## Choose the parameters of the simulation

```python
am= 1
pressure = 0. # use default value
pwv=4       # turn on or of the pwv
aer=0  
ozone=0.    # turn on or of the ozone
cloudext=0
```

```python
textstr = '\n'.join((
    f'airmass = {am:.1f}',
    f'pwv =  {pwv:.1f} mm',
    f'ozone = {ozone:.0f} DU'))
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
```

## Simulation

```python
all_trans = []
all_obs = []
for index, site in enumerate(List_Of_All_sites):
    all_obs.append(site)
    wl,atm=libsimulateVisible.ProcessSimulation(am,pwv,ozone,pressure,aer_num=aer,
                                                      prof_str='us',proc_str='sa',cloudext=cloudext,altitude=site)
    all_trans.append(atm)
```

## Plot

```python
fig,ax = plt.subplots()
for index in np.arange(NSites):   
    ax.semilogy(wl,all_trans[index],'-',color=hls_colors[index],label=all_obs[index])
ax.set_xlim(300,1000)
ax.grid()
ax.set_title(f"Total atmospheric transmission")
ax.set_xlabel("$\\lambda$ (nm)")
ax.set_ylabel("transmission")
ax.set_ylim(0.2,1.1)
ax.legend()
ax.text(0.05, 0.05, textstr, transform=ax.transAxes, fontsize=14,
        verticalalignment='bottom', bbox=props)
figname="multisitestotaltransmissionvssite_log.png"

plt.savefig(figname)
```

```python
fig,ax = plt.subplots()

for index in np.arange(NSites):   
    ax.plot(wl,all_trans[index],'-',color=hls_colors[index],label=all_obs[index])
ax.set_xlim(300,1200)
ax.grid()
ax.set_title(f"Total atmospheric transmission")

ax.set_xlabel("$\\lambda$ (nm)")
ax.set_ylabel("transmission")
ax.set_ylim(0.,1.1)
ax.legend(loc="lower right")
ax.text(0.05, 0.05, textstr, transform=ax.transAxes, fontsize=14,
        verticalalignment='bottom', bbox=props)
figname="multisitetotaltransmission_vssitelin.png"
plt.savefig(figname)
```

## Compare libradtran with standard pressure and libradtran with P=743

```python
wl1,atm1=libsimulateVisible.ProcessSimulation(am,pwv,ozone,pressure,aer_num=aer,
                                                      prof_str='us',proc_str='sa',cloudext=cloudext,altitude='LSST')

Pstand = Dict_Of_sitesPressures['LSST']
label1 = f"LSST site with P = {Pstand:.2f} hPa"


Ptest = 743.0
wl2,atm2=libsimulateVisible.ProcessSimulation(am,pwv,ozone,Ptest,aer_num=aer,
                                                      prof_str='us',proc_str='sa',cloudext=cloudext,altitude='LSST')
label2 = f"LSST site with P = {Ptest:.2f} hPa"

```

```python

plt.plot(wl1,atm1,'-r',label=label1)
plt.plot(wl2,atm2,'-b',label=label2)
plt.xlim(300,1200)
plt.grid()
plt.title(f"Total atmospheric transmission for airmass = {am} and 2 pressures")

plt.xlabel("$\\lambda$ (nm)")
plt.ylabel("transmission")
plt.ylim(0.4,1.05)
plt.legend()
figname="comparepressuretotaltransmission_lsst_lin.png"
plt.savefig(figname)
```

```python
the_leg = f"Total transmission ratio T(P = {Ptest:.1f} hPa)/T(P={Pstand:.1f} hPa)"
plt.plot(wl1,atm2/atm1,'-r',label=the_leg)

plt.xlim(300,1200)
plt.grid()
plt.title(f"Total atmospheric transmission ratio for airmass = {am}")

plt.xlabel("$\\lambda$ (nm)")
plt.ylabel("transmission")
plt.ylim(0.95,1.05)
plt.legend()
figname="comparepressuretotaltransmissionratio_lsst_lin.png"
plt.savefig(figname)
```
